```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import json

# Load JSON data
with open('../cryptos_preprocessed.json', 'r') as file:
    data = json.load(file)

# Convert to DataFrame
df = pd.DataFrame(data)


```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['current_price'], bins=50)
plt.title("Distribution of Current Prices")
plt.xlabel("Price (USD)")
plt.ylabel("Count")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Extraire uniquement les colonnes numériques
numeric_df = df[['current_price', 'market_cap', 'total_volume', 'price_change_24h_%']]

# Calcul des corrélations
correlation_matrix = numeric_df.corr()

# Affichage de la heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='YlGnBu', fmt=".2f")

plt.title("Corrélations entre les variables numériques")
plt.tight_layout()
plt.show()

```

```python
import requests
import pandas as pd
import matplotlib.pyplot as plt

# ID de la cryptomonnaie (par exemple 'bitcoin', 'ethereum')
crypto_id = 'bitcoin'

# Période de temps (par exemple, les 30 derniers jours)
days = 30

# URL pour récupérer les données historiques des prix
url = f'https://api.coingecko.com/api/v3/coins/{crypto_id}/market_chart?vs_currency=usd&days={days}'

# Requête GET à l'API
response = requests.get(url)

# Vérifier si la requête est réussie
if response.status_code == 200:
    data = response.json()
    # Extraire les prix et les timestamps
    prices = data['prices']
    
    # Convertir en DataFrame pandas
    df = pd.DataFrame(prices, columns=['timestamp', 'price'])
    df['timestamp'] = pd.to_datetime(df['timestamp'], unit='ms')  # Convertir les timestamps en datetime
    
    # Tracer l'évolution des prix
    plt.figure(figsize=(10, 6))
    plt.plot(df['timestamp'], df['price'], color='blue', label=f'{crypto_id.capitalize()} Price')
    plt.title(f'Evolution du prix de {crypto_id.capitalize()} au fil du temps')
    plt.xlabel('Date')
    plt.ylabel('Prix (USD)')
    plt.xticks(rotation=45)
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()
else:
    print("Erreur lors de la récupération des données.")
```
